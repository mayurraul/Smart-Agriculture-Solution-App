```python
import numpy as np
##import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
from math import ceil

mini_size=1024
num_epochs=25

def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        img_dict = pickle.load(fo, encoding='bytes')
    return img_dict

def load_dataset(file):
    file="mosei_senti_data.pkl"
    img_dict=unpickle(file)
##    file="batches.meta"
##    name_dict=unpickle(file)
##    label_names=name_dict["label_names".encode()]


    data=img_dict['data'.encode()]
    labels=img_dict['labels'.encode()]

    ##print("data 0 " ,data[0,0:1024])
    img_set=np.array(np.split(data,3,axis=1))
    img_reds=img_set[0]
    img_greens=img_set[1]
    img_blues=img_set[2]

    img_reds=np.reshape(img_reds,(10000,32,32))
    img_greens=np.reshape(img_greens,(10000,32,32))
    img_blues=np.reshape(img_blues,(10000,32,32))

    img_comp=np.concatenate((img_reds[:,:,:,np.newaxis],img_greens[:,:,:,np.newaxis],img_blues[:,:,:,np.newaxis]), axis=3, out=None)
##    print('img_comp',img_comp[10],img_comp[0])
##    print("img_red" ,img_comp.shape)

##    sam=img_comp[10000-1]
    ##sam=np.moveaxis(sam, 0, -1)
##    print("sam",sam.shape)
##    sam= cv2.resize(sam, (100, 100)) 
##    cv2.imshow("sample image",sam)

##    print("labels",label_names[labels[10000-1]])
    return( img_comp,labels)
print("Eager execution: {}".format(tf.executing_eagerly()))
X_train,Y_train=load_dataset('data_batch_1')
X_test,Y_test=load_dataset('test_batch')
X_train=np.array(X_train/255)
Y_train=np.array(Y_train)
X_test=np.array(X_test/255)
Y_test=np.array(Y_test)

m=X_train.shape[0]

print("loading dataset over")
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
print("creating model over")

model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
model.summary()
X_minis=[]
Y_minis=[]
t=ceil(10000/mini_size)
print("t",t)
for i in range(t-1):
    X_minis.append(X_train[i*mini_size:(i+1)*mini_size])
    Y_minis.append(Y_train[i*mini_size:(i+1)*mini_size])
    
X_minis.append(X_train[(t-1)*mini_size:])
Y_minis.append(Y_train[(t-1)*mini_size:])
X_minis=np.array(X_minis)
Y_minis=np.array(Y_minis)
print("Y_minis",Y_minis.ravel().shape)
for epoch in range(num_epochs):
    for i in range(t):
        history = model.fit(X_minis[i], Y_minis[i], epochs=1, validation_data=(X_test[0:1000,:,:,:], Y_test[0:1000]))

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()
test_loss, test_acc = model.evaluate(X_test,  Y_test, verbose=2)

print("test loss",test_loss,"test_acc",test_acc)

```
